```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

data = {
    'Payload': [2000, 2500, 1800, 3000, 2200, 2700, 3200, 1900, 2800, 2300],
    'Launch Site': ['Site A', 'Site B', 'Site A', 'Site C', 'Site B', 'Site A', 'Site C', 'Site B', 'Site A', 'Site C']
}

df = pd.DataFrame(data)

np.random.seed(42)
num_additional_payloads = 90
additional_data = {
    'Payload': np.random.randint(1500, 3500, num_additional_payloads),
    'Launch Site': np.random.choice(['Site A', 'Site B', 'Site C'], num_additional_payloads)
}

df_additional = pd.DataFrame(additional_data)
df = pd.concat([df, df_additional])

# Set style
sns.set(style="whitegrid")

# Scatter plot
plt.figure(figsize=(14, 8))
sns.scatterplot(x='Payload', y='Launch Site', data=df, s=150, hue='Launch Site', palette='viridis', edgecolor='w', alpha=0.8)

# Customizing the plot
plt.title('Payload vs. Launch Site', fontsize=18)
plt.xlabel('Payload', fontsize=14)
plt.ylabel('Launch Site', fontsize=14)
plt.legend(title='Launch Site', title_fontsize='14', loc='upper right')
plt.grid(True)

# Display the plot
plt.show()
```
