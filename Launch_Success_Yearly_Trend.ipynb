```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data = {
    'Launch Date': pd.to_datetime(['2010-01-01', '2010-02-01', '2011-01-01', '2011-02-01', '2012-01-01', '2012-02-01']),
    'Launch Success': [1, 0, 1, 1, 0, 1]
}

df = pd.DataFrame(data)

df['Year'] = df['Launch Date'].dt.year
yearly_success_rate = df.groupby('Year')['Launch Success'].mean()

sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))
sns.lineplot(x=yearly_success_rate.index, y=yearly_success_rate, marker='o', color='purple', linewidth=2, markersize=10)

plt.title('Yearly Average Launch Success Rate', fontsize=18)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Success Rate', fontsize=14)
plt.xticks(yearly_success_rate.index)
plt.grid(True)

plt.show()

```
