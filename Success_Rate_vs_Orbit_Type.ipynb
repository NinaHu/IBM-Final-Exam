```python
import matplotlib.pyplot as plt
import pandas as pd

data = {
    'Orbit Type': ['LEO', 'GTO', 'LEO', 'MEO', 'GTO', 'LEO', 'GTO', 'LEO', 'MEO', 'GTO'],
    'Success': [1, 0, 1, 1, 1, 0, 1, 1, 0, 1]
}

df = pd.DataFrame(data)

success_rate = df.groupby('Orbit Type')['Success'].mean().reset_index()

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(success_rate['Orbit Type'], success_rate['Success'], color='skyblue')
plt.title('Success Rate vs. Orbit Type', fontsize=16)
plt.xlabel('Orbit Type', fontsize=14)
plt.ylabel('Success Rate', fontsize=14)
plt.ylim(0, 1)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```
