```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv")
```

```python
#plt.figure(figsize=(200, 300))
# Define a custom color palette with dark red and dark blue
custom_palette = ["darkred", "darkblue", "darkblue"]

# Plotting the catplot with the custom color palette for 'Class'
sns.catplot(y="LaunchSite", x="FlightNumber", hue="Class", data=df, aspect=5, palette=custom_palette)
plt.xlabel("Flight Number", fontsize=25)
plt.ylabel("Launch Site", fontsize=25)
plt.show()
```
