```python
import folium
```

```python
# center of the map
latitude, longitude = 40.7128, -74.0060
```

```python
# basic map
my_map = folium.Map(location=[latitude, longitude], zoom_start=12)
my_map
```

```python
# map with location marker
marker_lat, marker_lng = 40.7128, -74.0060
folium.Marker(
    location=[marker_lat, marker_lng],
    popup='Marker Popup',
    icon=folium.Icon(color='blue')
).add_to(my_map)
my_map
```

```python
# map with a highlighted area (circle)
circle_lat, circle_lng = 40.7214, -74.0052
folium.Circle(
    location=[circle_lat, circle_lng],
    radius=500,  # in meters
    color='red',
    fill=True,
    fill_color='orange',
    fill_opacity=0.5
).add_to(my_map)
my_map
```

```python
# Connect dots with a line
line_lat1, line_lng1 = 40.7128, -74.0060
line_lat2, line_lng2 = 40.7220, -74.0055
line_lat3, line_lng3 = 40.7320, -74.0080

folium.PolyLine(
    locations=[(line_lat1, line_lng1), (line_lat2, line_lng2), (line_lat3, line_lng3)],
    color='green',
    weight=3
).add_to(my_map)
my_map
```
