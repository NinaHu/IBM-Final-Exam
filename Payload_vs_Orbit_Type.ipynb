```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

data = {
    'Payload': [2000, 2500, 1800, 3000, 2200, 2700, 3200, 1900, 2800, 2300],
    'Orbit Type': ['LEO', 'GTO', 'LEO', 'MEO', 'GTO', 'LEO', 'GTO', 'LEO', 'MEO', 'GTO']
}

df = pd.DataFrame(data)

sns.set(style="whitegrid")

orbit_colors = sns.color_palette('husl', n_colors=len(df['Orbit Type'].unique()))

plt.figure(figsize=(12, 8))
for i, orbit_type in enumerate(df['Orbit Type'].unique()):
    subset = df[df['Orbit Type'] == orbit_type]
    plt.scatter(subset['Payload'], subset['Orbit Type'], s=150, color=orbit_colors[i], alpha=0.7, edgecolor='w', label=orbit_type)

plt.title('Payload vs. Orbit Type', fontsize=18)
plt.xlabel('Payload', fontsize=14)
plt.ylabel('Orbit Type', fontsize=14)
plt.legend(title='Orbit Type', title_fontsize='14', loc='upper right')
plt.grid(True)

plt.show()
```
